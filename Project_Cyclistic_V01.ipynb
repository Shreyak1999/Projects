## CREATE SINGLE DATAFRAME

```python
Year_2023 = pd.concat([Jan_2023, Feb_2023, Mar_2023, Apr_2023, May_2023, Jun_2023, Jul_2023, Aug_2023, Sep_2023, Oct_2023, Nov_2023, Dec_2023], ignore_index=True)
```

## OBSERVE FIRST 5 AND LAST 5 DATA ENTRIES

```python
Year_2023
```

## DROP COLUMNS UNNECESSARY FOR ANALYSIS

```python
Year_2023 = Year_2023.drop(columns=['start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng'])
```

```python
Year_2023.tail()
```

## CHANGING DATA TYPE TO DATE

```python
Year_2023['started_at'] = Year_2023['started_at'].astype('datetime64[ns]')
Year_2023['ended_at'] = Year_2023['ended_at'].astype('datetime64[ns]')
```

## CREATING NEW VARIABLE

```python
Year_2023['ride_length'] = (Year_2023['ended_at'] - Year_2023['started_at'])/pd.Timedelta(minutes=1)
Year_2023['ride_length'] = Year_2023['ride_length'].astype('int32')
```

## DROPS ALL RIDE LENGTHS THAT ARE NEGATIVE OR LESS THAN 1 MINUTE. THE SECOND LINE RESETS THE INDEX TO DEFAULT 0,1,2,3...

```python
Year_2023 = Year_2023[Year_2023['ride_length'] >= 1]
Year_2023 = Year_2023.reset_index()
Year_2023 = Year_2023.drop(columns=['index'])
```

## CHANGING DATA TYPE

```python
Year_2023 = Year_2023.astype({'ride_id':'string', 'rideable_type':'category', 'member_casual':'category'})
Year_2023.info()
```

## REMOVES TRAILING AND LEADING WHITE SPACES IN THE COLUMN

```python
Year_2023['ride_id'] = Year_2023['ride_id'].str.strip()
```

## CREATE YEAR, MONTH AND WEEKDAY COLUMNS TO CORRESPOND TO VARIOUS DATES IN THE DATA

```python
from pandas import CategoricalDtype


Year_2023['year'] = Year_2023['started_at'].dt.year
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
Year_2023['month'] = Year_2023['started_at'].dt.month_name()
Year_2023['month'] = Year_2023['month'].astype(CategoricalDtype(categories=months, ordered=False))

weekday = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
Year_2023['day_of_week'] = Year_2023['started_at'].dt.day_name()
Year_2023['day_of_week'] = Year_2023['day_of_week'].astype(CategoricalDtype(categories=weekday, ordered=False))

Year_2023['hour'] = Year_2023['started_at'].dt.hour
Year_2023 = Year_2023.astype({'year':'int16', 'hour':'int8'})
```
